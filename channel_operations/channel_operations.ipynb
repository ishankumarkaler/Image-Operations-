```python
# import dependencies
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# reading image in BGR format using opencv2
img = cv2.imread("bird.png")
# shape of image
print(img.shape)
```

```python
# display the image 
plt.imshow(img);
```

```python
# create a new zero-initialized image array
img2 = np.zeros(img.shape).astype(int)
# change BGR channels to RGB channels
img2[:, :, 0] = img[:, :, 2]
img2[:, :, 1] = img[:, :, 1]
img2[:, :, 2] = img[:, :, 0]
```

```python
# display the changed image (RGB)
plt.imshow(img2);
```

```python
# displaying red, green, blue channels
fig = plt.figure(figsize = (20, 20))
colums = 3
rows = 1
# red channel
fig.add_subplot(rows, colums, 1)
plt.imshow(img2[:, :, 0]);
# green channel
fig.add_subplot(rows, colums, 2)
plt.imshow(img2[:, :, 1]);
# blue channel
fig.add_subplot(rows, colums, 3)
plt.imshow(img2[:, :, 2]);
```

```python
# binary thresholding colour channels
fig = plt.figure(figsize = (20, 20))
colums = 4
rows = 1
thresh_value = 60
fig.add_subplot(rows, colums, 1)
# red channel
plt.imshow((1 * (img2[:, :, 0] > thresh_value)).astype(float), cmap="gray");
fig.add_subplot(rows, colums, 2)
# green channel
plt.imshow((1 * (img2[:, :, 1] > thresh_value)).astype(float), cmap="gray");
fig.add_subplot(rows, colums, 3)
# blue channel
plt.imshow((1 * (img2[:, :, 2] > thresh_value)).astype(float), cmap="gray");
# all channels
fig.add_subplot(rows, colums, 4)
plt.imshow((1 * ((img2[:, :, 0] > thresh_value) + (img2[:, :, 1] > thresh_value) + (img2[:, :, 2] > thresh_value))).astype(float), cmap="gray");
```

```python
# reading a black-and-white image
imgg = cv2.imread("black.jpeg", 0)
plt.imshow(imgg, cmap = "gray");
```

```python
# applying colour filter to black-white image
imgg2 = np.zeros((imgg.shape[0], imgg.shape[1], 3)).astype(int)
imgg2[:, :, 0] = (0.8 * imgg).astype(int)
imgg2[:, :, 1] = (0.4 * imgg).astype(int)
imgg2[:, :, 2] = (0.3 * imgg).astype(int)
plt.imshow(imgg2);
```
